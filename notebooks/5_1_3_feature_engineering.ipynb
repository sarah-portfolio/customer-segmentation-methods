```python
def clean_and_calculate_iqr(df, column):
  """
  Cleans outliers from a specified column in a DataFrame using IQR and prints IQR values.

  Args:
    df: DataFrame containing the data.
    column: Name of the column to clean.

  Returns:
    A new DataFrame with outliers removed from the specified column.
  """
  df_clean = df.copy()
  Q1 = df_clean[column].quantile(0.25)
  Q3 = df_clean[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df_clean = df_clean[(df_clean[column] >= lower_bound) & (df_clean[column] <= upper_bound)]
  print(f"IQR for {column}: {IQR}")
  print(f"Lower bound for {column}: {lower_bound}")
  print(f"Upper bound for {column}: {upper_bound}")
  return df_clean

# Use the defined function to calculate iteratively the IQR of columns_to_clean:
columns_to_clean = ['Coolant temp', 'Lub oil pressure', 'Fuel pressure', 'lub oil temp', 'Coolant pressure']
cleaned_dfs = {}
for column in columns_to_clean:
    cleaned_df_name = f"df_clean_{column.replace(' ', '_').lower()}"
    cleaned_dfs[cleaned_df_name] = clean_and_calculate_iqr(df, column)
```

```python
# Clean each column and store the results in separate DataFrames
df_clean_lop = clean_and_calculate_iqr(df, 'Lub oil pressure')
df_clean_fp = clean_and_calculate_iqr(df, 'Fuel pressure')
df_clean_cp = clean_and_calculate_iqr(df, 'Coolant pressure')
df_clean_lot = clean_and_calculate_iqr(df, 'lub oil temp')
df_clean_ct = clean_and_calculate_iqr(df, 'Coolant temp')
```

```python
def plot_raw_and_clean_linreg(df, df_clean, col1, col2, col3, col4, col5):
  """
  Plots raw and cleaned linear regression models for two pairs of columns.

  Args:
    df: DataFrame containing the raw data.
    df_clean: DataFrame containing the cleaned data.

  """

  # Plot for the first pair of columns
  plt.figure(figsize=(15, 7))

  # With outliers.
  plt.subplot(1, 2, 1)
  plt.scatter(df[col1], df[col2], label='Data')

  lr_raw = LinearRegression()
  lr_raw.fit(df[[col1]], df[col2])
  y_pred_raw = lr_raw.predict(df[[col1]])
  plt.plot(df[col1], y_pred_raw, color='red', label='Line of best fit')

  # Without outliers.
  plt.subplot(1, 2, 2)
  plt.scatter(df_clean[col1], df_clean[col2], label='Data')

  lr_clean = LinearRegression()
  lr_clean.fit(df_clean[[col1]], df_clean[col2])
  y_pred_clean = lr_clean.predict(df_clean[[col1]])
  plt.plot(df_clean[col1], y_pred_clean, color='red', label='Line of Best Fit')

  mse_clean = mean_squared_error(df_clean[col2], y_pred_clean)
  r2_clean = r2_score(df_clean[col2], y_pred_clean)
  plt.title('Cleaned data')
  plt.legend()
  plt.text(df_clean[col1].min(), df_clean[col2].min(), f'MSE: {mse_clean:.2f}\nR-squared: {r2_clean:.2f}', fontsize=12)

  plt.show()
```
